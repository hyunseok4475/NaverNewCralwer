```python
for j in range(0,len(step_100_day)):
    start_day_year = str(step_100_day[j].year)
    if(step_100_day[j].month < 10):
        start_day_month = str(step_100_day[j].month)
        start_day_month = '0'+start_day_month
    else:
        start_day_month = str(step_100_day[j].month)
    
    if(step_100_day[j].day < 10):
        start_day_day = str(step_100_day[j].day+1)
        start_day_day = '0'+start_day_day
    else:
        start_day_day = str(step_100_day[j].day+1)

    end_day_year = str(step_100_day[j+1].year)

    if(step_100_day[j+1].month < 10):
        end_day_month = str(step_100_day[j+1].month)
        end_day_month = '0'+end_day_month
    else:
        end_day_month = str(step_100_day[j+1].month)

    if(step_100_day[j+1].day < 10):
        end_day_day = str(step_100_day[j+1].day)
        end_day_day = '0'+end_day_day
    else:
        end_day_day = str(step_100_day[j+1].day)


    start_day = start_day_year+"."+start_day_month+"."+start_day_day
    no_dot_start_day = start_day_year+start_day_month+start_day_day
    end_day = end_day_year+"."+end_day_month+"."+end_day_day
    no_dot_end_day = end_day_year+end_day_month+end_day_day

    naver_link = []
    for i in range(1,4000,10):
        headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Whale/2.7.99.20 Safari/537.36'}
    #start_day ~ end_day '2018.05.21 ~ 2020.05.21' 같이 표기함
    #query , start_day , end_day, start_day에 . 없는거, end_day에 .없는거, 시작기사 번호
        url = "https://search.naver.com/search.naver?&where=news&query={}D&sm=tab_pge&sort=0&photo=0&field=1&reporter_article=&pd=3&ds={}&de={}&docid=&nso=so:r,p:from{}to{},a:t&mynews=0&cluster_rank=29&start={}&refresh_start=0".format(parse.quote(keyword),start_day, end_day,no_dot_start_day,no_dot_end_day,i)
        html = requests.get(url,headers=headers).text
        bsObject = BeautifulSoup(html, "html.parser")  
        links = bsObject.find_all("a",{"class":"_sp_each_url"})
        link_group = []
        for link in links:
            link_group.append(link['href'])
        link_temp = [k for k in link_group if 'news.naver' in k]
        naver_link.extend(link_temp)
        print(link_temp)
        print(i)
    del(html, link, link_group, link_temp)
    naver_link = list(set(naver_link))
    title = []
    date  = []
    press = []
    contents = []
    newslink = []
    notlink = []
    for link in naver_link:
        try:
            news = urlopen(link)
            news = BeautifulSoup(news.read(), "html.parser")
            
            # 신문사
            tmp = news.select('#footer address')[0].a.get_text()
        
            # 내용
            text = news.select('#articleBodyContents')[0].get_text().replace('\n', " ")
            btext = text.replace("// flash 오류를 우회하기 위한 함수 추가 function _flash_removeCallback() {}", "")
            
            
            press.append(tmp) 
            contents.append(btext.strip())  
            
            del(tmp, text, btext)
        
        
            # 날짜
            tmp = news.find('span', {'class': 't11'}).text
            tmp = re.findall("\d+.\d+.\d+", str(tmp))
            date.extend(tmp) # 날짜 
            
            # 제목
            title.append(news.title.text)     
            
            # 링크
            newslink.append(link)
            
            del(tmp)
            
        except:
            notlink.append(link)
    print("전체 뉴스 링크 수: ", len(naver_link))
    print("크롤링하지 못한 링크 수: ", len(notlink))
    print("제목수: ", len(title))
    print("날짜수: ",len(date))
    print("출판사수: ",len(press))
    print("내용수: ",len(contents))
    print("뉴스링크수: ",len(newslink))

    notlink2 = []

    for link in notlink:
        try:
            news = urlopen(link)
            news = BeautifulSoup(news.read(), "html.parser")
        
            # 내용
            text = news.find('div', {'class':'naver_post'}).get_text()
            contents.append(text)  
        
            # 신문사
            tmp = news.find('span', {'class':'logo'}).img.get('alt')
            press.append(tmp) 
            
            # 날짜
            tmp = news.find('div', {'class': 'info'}).span.get_text()
            tmp = re.findall("\d+.\d+.\d+", str(tmp))
            date.extend(tmp) # 날짜 
            
            
            # 제목
            title.append(news.title.text)     
            
            # 링크
            newslink.append(link)
            
            del(news, tmp, text)
            
        except:
            notlink2.append(link)

    print("1. 전체 네이버 뉴스 기사 수: ", len(naver_link))
    print("2. 크롤링 된 뉴스 기사 수: ", len(naver_link) - len(notlink))
    print("3. 크롤링 되지 않은 뉴스 기사 수: ", len(notlink))
    print("4. 위의 기사 중 스포츠 뉴스 기사 수: ", len(notlink) - len(notlink2))
    print("5. 스포츠 뉴스 크롤링 방법을 통해서도 크롤링 되지 않은 최종 뉴스 기사 수: ", len(notlink2))
    print(len(title))
    print(len(date))
    print(len(press))
    print(len(contents))
    print(len(newslink))


    naver_news = pd.DataFrame({"title": title,
                            "date": date,
                            "press": press,
                            "contents": contents,
                            "link": newslink})

    naver_news.to_csv("Data/naver_news_"+no_dot_start_day+"~"+no_dot_end_day+".csv", encoding = 'utf-8', index = False)
```
